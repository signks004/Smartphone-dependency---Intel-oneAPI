```python
sns.set_style("whitegrid")
plt.figure(figsize = (8,5))
db = db.rename(columns = {"Do you feel anxious when you  don't have your smartphones.":"Do you feel anxious when you  dont have your smartphones."})
eda_percentage = db['Do you feel anxious when you  dont have your smartphones.'].value_counts(normalize = True).rename_axis('Do you feel anxious when you  dont have your smartphones.').reset_index(name = 'Percentage')

ax = sns.barplot(x = 'Do you feel anxious when you  dont have your smartphones.', y = 'Percentage', data = eda_percentage.head(10), palette='Purples')
for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height:.0%}', (x + width/2, y + height*1.02), ha='center', fontweight='bold')
```
